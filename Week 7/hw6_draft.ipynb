```python
#hw6_1
def heart(im):
    im_new = im.copy()
    
    col,row,layer = im.shape #unpack dimensions 
    print col, row, layer
    row = float(row)
    col = float(col)

    y,x = np.ogrid[0:col, 0:row]  #x in charge of hori, y in charge of vertical
    print x.shape, y.shape
    y = np.array(y,dtype='float') 
    x = np.array(x,dtype='float') 

    mask1 = (x<=row*1/2) & (y>=col*1/3) & (y>= (4*col/(3*row))*x+col/3)
    mask2 = (x>=row*1/2) & (y>=col*1/3) & (y>= (-4*col/(3*row))*x+5*col/3)
    mask3 = (x<=row*1/2) & (y<=col*1/3) & ((x-row/4)**2 + (y-col/3)**2 >= (row**2)/16)
    mask4 = (x>=row*1/2) & (y<=col*1/3) & ((x-3*row/4)**2 + (y-col/3)**2 >= (row**2)/16)
    mask = mask1 | mask2 | mask3 | mask4
    im_new[mask] = [255,189,234] # mask certain pixel to be pink
    plt.imshow(im_new)
    plt.show()
```

```python
#hw6_2
def uni_blur(im,k): #im already grey-scale
    im_new = im[:,:,0].copy() #now im_new is 2-d array
    print im_new.shape
    print type(im_new)
    im_process = im[:,:,0].copy() #now im_process is 2-d array
    row,col,layer = im.shape #unpack dimensions 
    
    print row, col, layer #row number means length of y axis, col num means length of x axis
    for i in range(1,row-1):
        for j in range(1,col-1):
            #print "(",i,',',j,')'
            sub_square = im_new[(i-(k-1)/2):(i+(k-1)/2)+1,(j-(k-1)/2):(j+(k-1)/2)+1]
            #print sub_square.shape
            sum1 = float(np.sum(sub_square))
            #print sum1
            new_value = sum1/(k**2)
            #print new_value
            im_process[i,j] = new_value
    
    im_result = [[[im_process[i,j]]*3 for j in range(col)] for i in range(row)]
    print np.shape(im_result) #im_result is a python list, not nparray, so we use np.shape() to get shape
    plt.imshow(im_result)
    plt.show()
    
    #sub_square = im_new[(i-(k-1)/2):(i+(k-1)/2)+1),(j-(k-1)/2):(j+(k-1)/2)+1)]

def Gaussian_blur(im,k):
    im_new = im[:,:,0].copy() #now im_new is 2-d array
    print im_new.shape
    im_process = im[:,:,0].copy() #now im_process is 2-d array
    row,col,layer = im.shape #unpack dimensions 
    
    sigma = 1# the dimension of filter is specified by k already
    
    filter=np.array([[0]*k]*k,dtype='float')
    print filter
    for x in range(k):
        for y in range(k):
            filter[x,y]=np.exp(-((x-(k-1)*0.5)**2+(y-(k-1)*0.5)**2)/(2.0*sigma**2))
    filter_sum=np.sum(filter)
    filter=filter/filter_sum  #now filter is ready to use  
    
    print row, col, layer #row number means length of y axis, col num means length of x axis
    for i in range(1,row-1):
        for j in range(1,col-1):
            #print "(",i,',',j,')'
            sub_square = im_new[(i-(k-1)/2):(i+(k-1)/2)+1,(j-(k-1)/2):(j+(k-1)/2)+1]
            #print sub_square.shape
            sub_square = sub_square * filter
            sum1 = float(np.sum(sub_square))
            #print sum1
            im_process[i,j] = sum1
    
    im_result = [[[im_process[i,j]]*3 for j in range(col)] for i in range(row)]
    print np.shape(im_result) #im_result is a python list, not nparray, so we use np.shape() to get shape
    plt.imshow(im_result)
    plt.show()
    


def blurring(im, method):
    if (method == "uniform"):
        print "use uniform"
        uni_blur(im,3)
    elif (method == "Gaussian"):
        print "use Gaussian"
        Gaussian_blur(im,3)
    
```

```python
img=mpimg.imread('kitty-cat.jpg')
#blurring(img,"Gaussian")
def salt_pepper(im,ps=.1,pp=.1): 
    im1=im[:,:,0].copy() #2 dimension im1, from taking the 0th layer of color, Red
    n,m=im1.shape #n is No of row, m is No of columns
    for i in range(n):
        for j in range(m):
            b=np.random.uniform() #b is randomly from 0 to 1
            if b<ps:
                im1[i,j]=1
            elif b>1-pp:
                im1[i,j]=0
    noisy_im = np.repeat(im1[:, :, np.newaxis], 3, axis=2)
    #noisy_im=[[[im1[i,j]]*3 for j in range(m)] for i in range(n)] #im1[i,j] is the Red_number of each pixel, after
        # adding noise. im1[i,j]*3 will generate a list of 3 elements with data be im1[i,j], to get back 3-layer style                                                                
    return noisy_im
```

```python
#hw6_3
def detect_edge(im, method):
    k=3
    im = im/255.0
    im_new = im[:,:,0].copy() #now im_new is 2-d array
    print im_new.shape
    im_process = im[:,:,0].copy() #now im_process is 2-d array
    row,col,layer = im.shape #unpack dimensions 
    
    print row, col, layer #row number means length of y axis, col num means length of x axis
    
    vertical_filter=np.array([[-1,0,1],
                          [-2,0,2],
                          [-1,0,1]])
    horizontal_filter=np.array([[-1,-2,-1],
                            [0,0,0],
                            [1,2,1]])
    for i in range(1,row-1):
        for j in range(1,col-1):
            #print "(",i,',',j,')'
            sub_square = im_new[(i-(k-1)/2):(i+(k-1)/2)+1,(j-(k-1)/2):(j+(k-1)/2)+1]
            #print sub_square.shape
            if (method == "vertical"):
                sub_square = sub_square * vertical_filter
                Sv = (np.sum(sub_square)+4)/8
                #print sum1
                im_process[i,j] = Sv
            elif (method == "horizontal"):
                sub_square = sub_square * horizontal_filter
                Sh = (np.sum(sub_square)+4)/8
                #print sum1
                im_process[i,j] = Sh
            elif (method == "both"):
                sub_square1 = sub_square * vertical_filter
                sub_square2 = sub_square * horizontal_filter
                Sv = (np.sum(sub_square1)+4)/8
                Sh = (np.sum(sub_square2)+4)/8
                S = ((Sv**2+Sh**2)**1/2)/4
                im_process[i,j] = S
                
    
    im_result = [[[im_process[i,j]]*3 for j in range(col)] for i in range(row)]
    print np.shape(im_result) #im_result is a python list, not nparray, so we use np.shape() to get shape
    plt.imshow(im_result)
    plt.show()
    #img = img/255.0 #Convert to 64-bit floating point.
```
